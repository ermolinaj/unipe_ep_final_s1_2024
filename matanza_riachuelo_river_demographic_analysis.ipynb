```python
import pandas as pd
import matplotlib.pyplot as plt
```

# Importación de los datasets de Acumar

Se importan para visualizar individualmente los datasets obtenidos

```python
tons_of_waste_15_19 = pd.read_csv("datasets/water_mirror_cleaning/toneladas_retiradas_espejo_2015_2019.csv", sep=';', encoding='latin1')

tons_of_waste_15_19
```

```python
tons_of_waste_19_20 = pd.read_csv("datasets/water_mirror_cleaning/toneladas_retiradas_espejo_2019_2020.csv", sep=',', encoding='latin1')

tons_of_waste_19_20
```

```python
tons_of_waste_20_21 = pd.read_csv("datasets/water_mirror_cleaning/toneladas_retiradas_espejo_2020_2021.csv", sep=',', encoding='latin1')

tons_of_waste_20_21
```

```python
tons_of_waste_22 = pd.read_csv("datasets/water_mirror_cleaning/toneladas_retiradas_espejo_2022.csv", sep=";", encoding='latin1')

tons_of_waste_22
```

```python
tons_of_waste_23 = pd.read_csv("datasets/water_mirror_cleaning/toneladas_retiradas_espejo_2023.csv", sep=";", encoding='latin1')

tons_of_waste_23
```

# Consolidación de todos los datasets importados

Integro todos los datasets en uno solo para poder utilizarlo con mayor facilidad

```python
tons_of_waste = tons_of_waste_15_19

previously_red_dataframes = [tons_of_waste_19_20, tons_of_waste_20_21, tons_of_waste_22, tons_of_waste_23]

for df in previously_red_dataframes:
    tons_of_waste = pd.concat([tons_of_waste, df], ignore_index=True)
    
tons_of_waste
```

# Representación gráfica de los datasets

## Evolución temporal de la limpieza de la cuenca por año

```python
tons_of_waste_cleansed_by_year = tons_of_waste.groupby("anio")[["toneladas_retiradas"]].sum()
tons_of_waste_cleansed_by_year
```

```python
# tons_of_waste_cleansed_by_year.plot()
tons_of_waste_cleansed_by_year.plot()
ax = plt.gca()
# ax..set_title("Año")
ax.set_ylim([0, 7000])
```

## Limpieza acumulada de la cuenca

```python
accumulated_tons_of_waste_cleansed_by_year = tons_of_waste_cleansed_by_year.copy()

for i in range(accumulated_tons_of_waste_cleansed_by_year.size):
    if i > 0:
        accumulated_tons_of_waste_cleansed_by_year.iloc[i] = accumulated_tons_of_waste_cleansed_by_year.iloc[i] + accumulated_tons_of_waste_cleansed_by_year.iloc[i-1]

tons_of_waste_cleansed_by_year
```

```python
accumulated_tons_of_waste_cleansed_by_year.plot()
ax = plt.gca()
ax.set_ylim([0, 40000])

print(f"Total toneladas Retiradas: {accumulated_tons_of_waste_cleansed_by_year.iloc[-1].toneladas_retiradas}")
```
